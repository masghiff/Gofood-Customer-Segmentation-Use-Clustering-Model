```python
data = df.groupby(['merchant_area', 'isDiscount']).agg({'isDiscount' : 'count'})
percentage = df.groupby(['merchant_area']).agg({'isDiscount':'count'})
data.div(percentage, level='merchant_area') * 100
plt.rcParams['figure.figsize'] = (10,7)
ax = data.plot(kind='barh')
for i in ax.containers:
  ax.bar_label(i,)
```

```python
discprice = df[df.discount_price.notnull()]
discprice['perc_disc'] = 100*(discprice['price']-discprice['discount_price'])/discprice['price']
plt.rcParams["figure.figsize"] = (10,7)
ax = discprice.groupby('merchant_area').mean()['perc_disc'].plot(kind='barh')
plt.title("Rata-Rata Persentase Discount Setiap Kota", fontsize = 20)
plt.xlabel('')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
for i in ax.containers:
  ax.bar_label(i,)
```

```python
disc = pd.DataFrame()
col = discprice['merchant_area'].unique()

for i in col:
  data = discprice[discprice.merchant_area==i]
  data[['brand','lokasi']] = data['merchant_name'].str.split(',',expand=True, n=2)
  data = pd.DataFrame(data.groupby('brand').mean()['perc_disc'].sort_values(ascending=False).head(10).reset_index())
  data = pd.concat([disc,data]).head(10)
  print('Rata-Rata Persentase Diskon Brand Kota {}\n'.format(i),data)
```

```python
cat1 = data_category['category1'].unique().tolist()
cat2 = data_category['category2'].unique().tolist()
cat3 = data_category['category3'].unique().tolist()
cat4 = data_category['category4'].unique().tolist()
cat5 = data_category['category5'].unique().tolist()
cat = cat1 + cat2 + cat3 + cat4 + cat5
cat = list(set(cat))
cat = [x for x in cat if x is not None]
print(cat)
```

```python
city = df['merchant_area'].unique()
kota = []
kategori = []
jmlh_merchant = []

for i in city:
  for j in cat:
    data_city = df[df.merchant_area==i]
    n = data_city[data_city.category.str.startswith(j)].drop_duplicates('merchant_name').shape[0]
    kota.append(i)
    kategori.append(j)
    jmlh_merchant.append(n)
    
data = pd.DataFrame(list(zip(kota,kategori,jmlh_merchant)),columns=['Kota','Kategori','Jumlah_Merchant'])
jkt = data[data.Kota=='jakarta'].sort_values(['Jumlah_Merchant'],ascending=False).head(5)
display('Jakarta Most Category=', jkt)
print('\n')
mdn = data[data.Kota=='medan'].sort_values(['Jumlah_Merchant'],ascending=False).head(5)
display('Medan Most Category=', mdn)
print('\n')
sby = data[data.Kota=='surabaya'].sort_values(['Jumlah_Merchant'],ascending=False).head(5)
display('Surabaya Most Category=', sby)
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for i in object_cols:
    df[i]=df[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
from sklearn.preprocessing import StandardScaler
dc = df.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
#Scaling
scaler = StandardScaler()
scaler.fit(dc)
scaled_dc = pd.DataFrame(scaler.transform(dc),columns= dc.columns )
print("All features are now scaled")
```

```python
print("Dataframe to be used for further modelling:")
scaled_dc.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled_dc)
PCA_dc = pd.DataFrame(pca.transform(scaled_dc), columns=(["PC 1", "PC 2","PC 3"]))
PCA_dc.describe().T
```

```python
x =PCA_dc["PC 1"]
y =PCA_dc["PC 2"]
z =PCA_dc["PC 3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import warnings
import sys
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_dc)
Elbow_M.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
AC = AgglomerativeClustering(n_clusters=4)
# fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_dc)
PCA_dc["Clusters"] = yhat_AC
#Adding the Clusters feature to the orignal dataframe.
df["Clusters"]= yhat_AC
```
